# COADS Trends
This notebook assesses trends in precipitation from the COADS data and produces visualizations of those trends. The data analyzed arises from the composited precipitation fractions on a $1^\circ \times 1^\circ$ grid.

```python
def latlon_from_ij(ij):    
    """Return the [(latitude, longitude)] based on the 1x2 list [ij] with indices i and j, assuming that
    the data is on a 1x1 degree grid."""
    i = ij[0]
    j = ij[1]
    lat = i-90 + 0.5
    lon = j  + 0.5
    if lon > 180.0: lon -= 360.
    return lat,lon
```

```python
start_yr = 1950
end_yr = 2019
years = np.arange(start_yr, end_yr + 1) # List of years
x = years - np.average(years)

def trend(y, w, verbose=False):
    """Compute and return information regarding trends in precipitation fraction timeseries [y] over the
    70-year period of record between 1950 and 2019. [verbose] is [True] if information is to be printed."""
    X = sm.add_constant(x)
    mod_wls = sm.WLS(y, X) # Unweighted best fit
    res_wls = mod_wls.fit()

    intercept = res_wls.params[0]
    slope = res_wls.params[1]
    pvalue = res_wls.pvalues[1]
    confint = res_wls.conf_int()[1]
    
    percent_per_decade_absolute = 1000*slope
    percent_per_decade_relative = percent_per_decade_absolute/intercept
    
    ######## TEST FOR SIGNIFICANT TREND #########
    
    N = res_wls.nobs # Number of observations (after dropping)
    rho1 = sm.tsa.acf(y, missing='conservative')[1] # Lag-1 autocorrelation
    N_eff = N * (1 - rho1)/(1 + rho1) # Effective sample size
    
    fit = slope * x + intercept # Fit from regression
    y_err = np.nansum((y - fit)**2)
    x_err = np.nansum((x - x.mean()) ** 2)
    se = ((1 / (N_eff - 2)) * (y_err / x_err))**0.5 # Calculate std. err.
    
    conf = 0.95
    conf_adj = (1 + conf) / 2
    t_crit = stats.t.ppf(conf_adj, N_eff - 2)
    t = slope / se
    
    confint = np.array([slope - se * t_crit, slope + se * t_crit])
    sig = np.abs(t) > np.abs(t_crit) # Significant if t > t_crit
    
    ##########################################

    if verbose:
        print(res_wls.summary())
        
        print('intercept = %5.3f%%' % (100*intercept))
        print('slope = %5.3f%% (absolute) per decade' % (percent_per_decade_absolute))  
        print('slope = %5.3f%% (relative) per decade' % (percent_per_decade_relative)) 
        print('pvalue = ',pvalue)
        print('Conf. Int. = %5.3f , %5.3f' % (1000*confint[0], 1000*confint[1]))

        print()
        print('AUTOCORRELATION-ADJUSTED SIGNIFICANCE TESTING')
        print('# obs: %d'%N)
        print('# adj: %.2f'%N_eff)
        print('crit-t: %.3f'%t_crit)
        print('t: %.3f'%t)
        
        if sig:
            print('Significant at 95% level')
        else:
            print('Not significant at 95% level')
            
    results = {'trenda' : percent_per_decade_absolute, 'trendr' : percent_per_decade_relative, 
               'slope' : slope, 'intercept' : intercept , 'sig' : sig}
            
    return results
```

```python
trends = {}

for k in composites.keys():
    
    trends[k] = {}
        
    print(k)
    f = composites[k]['f']
    s = composites[k]['s']
    sh = f.shape
    
    for iseas in range(len(seasons)):
        seas = seasons[iseas]
        trends[k][seas] = {}
        d = trends[k][seas]
        print(seas)
        trenda =   np.zeros(sh[2:], dtype='float32')
        trendr =   np.zeros(sh[2:], dtype='float32')
        mean  =   np.zeros(sh[2:], dtype='float32')
        sig   =   np.zeros(sh[2:], dtype='int8')

        for ilat in range(180):
            for ilon in range(360):
    
                y = f[:, iseas, ilat,ilon]
                w = s[:, iseas, ilat,ilon]
            
                nans = np.sum(np.isnan(y))
                if nans <= 5: # LIMIT YEARS OF MISSING DATA
                    results = trend(y,w)        
                    trenda[ilat,ilon] = results['trenda']
                    trendr[ilat,ilon] = results['trendr']
                    mean[ilat,ilon] = results['intercept']
                    sig[ilat,ilon] = results['sig']
                else:
                    trenda[ilat,ilon] = np.nan
                    trendr[ilat,ilon] = np.nan
                    mean[ilat,ilon] = np.nan
                    sig[ilat,ilon] = False
        
        trenda[1-ls_mask] = np.nan
        trendr[1-ls_mask] = np.nan
        mean[1-ls_mask] = np.nan
        
        d['trenda'] = trenda   
        d['trendr'] = trendr   
        d['mean'] = mean   
        d['sig'] = sig   
```

```python
with open(join(data_dir, 'coads_1950-2019_filtered.pkl'), 'rb') as f:
    df = pickle.load(f)  
```

```python
start_yr = 1950
end_yr = 2019

nyrs = end_yr - start_yr + 1
nreports_all = np.zeros((6, nyrs, 12, 36, 72), dtype='int32')
nprecip_all  = np.zeros((6, nyrs, 12, 36, 72), dtype='int32')

for iyr in range(start_yr, end_yr + 1):
    print(iyr)
    ioffset = iyr-start_yr
    
    dfsub0 = df[df.YR == iyr]

    for imo in range(12):
        dfsub = dfsub0[dfsub0.MO == imo+1]
 
        for i in range(36):
            lat0 = i*5.0 - 90.0
            lat1 = lat0 + 5.0
        
            dlatsub = dfsub[dfsub.LAT >= lat0]
            dlatsub = dlatsub[dlatsub.LAT < lat1]
        
            for j in range(72):
                lon0 = j*5.0
                lon1 = lon0 + 5.0
            
                dlonsub = dlatsub[dlatsub.LON >= lon0]
                dlonsub = dlonsub[dlonsub.LON < lon1]
                dprecip = dlonsub[dlonsub.WW >= 50]
            
                for ipt in range(6):
                    nreports_all[ipt,ioffset,imo, i,j] = np.sum(dlonsub['PT'] == ipt)
                    nprecip_all[ipt,ioffset,imo, i,j] = np.sum(dprecip['PT'] == ipt)
                
nreports_all = np.flip(nreports_all, axis=3)
nprecip_all = np.flip(nprecip_all, axis=3)

del df
```

```python
ipts = [0, 5]   # Use only navy ships and 'ship'

nreports = np.sum(nreports_all[ipts], axis=0)
nprecip = np.sum(nprecip_all[ipts], axis=0)

nreports70 = np.sum(nreports, axis=(0,1))
nprecip70 = np.sum(nprecip, axis=(0,1))
```

```python
def plot_map(data, title=None, bottom=0.0, height=0.3, vmin=None, vmax=None,
            boxes=[], boxcolor='black', extend='both', cmap='None', 
            colorbartitle=None, colorbarticks=None, colorbarticklabels=None, cbtickrot=None,
            xlabels=True,ylabels=True,projection = ccrs.PlateCarree(central_longitude=180), ax=None):

    left = 0.15
    width = 0.7

    central_lon = 180.0
    projection = projection

    rect = [left, bottom, width, height]
    if not ax:
        ax = plt.axes(rect, projection=projection)
    
    img_extent = (-180, 180.0, -90, 90)
    glonticks = list(range(-150,190,30))
    glatticks = list(range(-90,100,30))
    
    im = ax.imshow(data, origin='upper', extent=img_extent, 
          transform=projection, cmap=plt.get_cmap(cmap), vmin=vmin, vmax=vmax)

    lw = 1
    lc = 'gray'
### BEGIN BUG WORKAROUND
 # Note weird displacement of longitude
    ax.plot([0., 0.],[-80., 90], linewidth=lw, linestyle='--', color=lc)
    ax.plot([0., 30],[-60., -60], linewidth=lw, linestyle='--', color=lc)
    ax.plot([0., 30],[-30., -30], linewidth=lw, linestyle='--', color=lc)
    ax.plot([0., 30],[0., 0], linewidth=lw, linestyle='--', color=lc)
    ax.plot([0., 30],[30., 30], linewidth=lw, linestyle='--', color=lc)
    ax.plot([0., 30],[60., 60], linewidth=lw, linestyle='--', color=lc)
### END WORKAROUND

    gl = ax.gridlines(xlocs=np.arange(-180,181, 30), ylocs=np.arange(-90,91,30),
                  linewidth=lw, linestyle='--', color=lc,draw_labels=True)
    gl.top_labels = False
    gl.right_labels = False
    
    if not xlabels:
        gl.bottom_labels = False
    if not ylabels:
        gl.left_labels = False


    gl.xlocator = mticker.FixedLocator(glonticks)
    gl.ylocator = mticker.FixedLocator(glatticks)
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
 
    ax.set_title(title, loc='left', fontsize=14)
    ax.coastlines()
    ax.add_feature(cartopy.feature.OCEAN, color='lightgray')
    ax.add_feature(cartopy.feature.LAND, zorder=100, edgecolor='black')


    for b in boxes:
        lat0 = b[0][0]
        lat1 = b[1][0]
        lon0 = b[0][1]
        lon1 = b[1][1]
        
        yvec = np.array([lat0, lat1, lat1, lat0, lat0])
        xvec = np.array([lon0, lon0, lon1, lon1, lon0]) + central_lon
        ax.plot(xvec, yvec, linestyle='-', color='black', linewidth=2)
        ax.plot(xvec, yvec, linestyle='dotted', color='white', linewidth=2)
                    
    cbar = plt.colorbar(mappable=im, cax=None, ax=None, shrink=0.6, pad=0.02,
            extend=extend)

    if colorbarticks != None:
        cbar.set_ticks(colorbarticks)
        if colorbarticklabels != None:
            if cbtickrot:
                cbar.set_ticklabels(colorbarticklabels)
            else:
                cbar.set_ticklabels(colorbarticklabels, rotation='vertical')
                for t in cbar.ax.get_yticklabels():
                    t.set_verticalalignment('bottom') 
    if colorbartitle != None:
        cbar.set_label(colorbartitle,size=10)
```

```python
iseason = 4 # Select season out of [DJF, MAM, JJA, SON, Annual] by index
year = np.array(range(start_yr,end_yr+1))

fig = plt.figure(figsize=(16,16))
figlab = ['a)', 'b)', 'c)', 'd)','e)','f)','g)','h)','i)']

timeseries = []

# For each region
for ibox, b in enumerate(boxes):
    subfigure = plt.subplot(3, 3, ibox+1)
    
    # Determine lat/lon extents of boxes and their indices in the data
    lat0 = b[0][0]
    lat1 = b[1][0]
    lon0 = b[0][1]
    lon1 = b[1][1]
    
    i2 = 36-int((lat0+90.1)/5)
    i1 = 36-int((lat1+90.1)/5)
    j1 = int((lon0 + 360.1)/5)
    j2 = int((lon1 + 360.1)/5)
    
    
    s0 = "%.0f" % np.abs(lat0)
    if (lat0 > 0): s0 = s0 + "N"
    if (lat0 < 0): s0 = s0 + "S"
    s1 = "%.0f" % np.abs(lat1)
    if (lat1 > 0): s1 = s1 + "N"
    if (lat1 < 0): s1 = s1 + "S"
    
    if lon0 > 180: lon0 -= 360
    if lon1 > 180: lon1 -= 360
    if lon0 < -180: lon0 += 360
    if lon1 < -180: lon1 += 360
        
    s2 = "%.0f" % np.abs(lon0)
    if (lon0 > 0): s2 = s2 + "E"
    if (lon0 < 0): s2 = s2 + "W"
    s3 = "%.0f" % np.abs(lon1)
    if (lon1 > 0): s3 = s3 + "E"
    if (lon1 < 0): s3 = s3 + "W"
    
    # Tally up counts within the regions
    nseries = np.zeros(nreports.shape[0])
    pseries = np.zeros(nprecip.shape[0])
    for imo in months[iseason]:
        nseries += np.sum(nreports[:, imo, i1:i2, j1:j2], axis=(1,2))
        pseries += np.sum(nprecip[:, imo, i1:i2, j1:j2], axis=(1,2))
    ratio = pseries/nseries
    stderr = np.sqrt(pseries)/nseries
    stderr = np.array([pair[1] for pair in [MNcalc(int(p), int(n)) for p, n in zip(pseries, nseries)]])

    ####### UNWEIGHTED FIT ########
    model = sm.WLS(ratio, sm.add_constant(year))
    res_wls = model.fit()
    
    b1 = res_wls.params[1]
    b0 = res_wls.params[0]
    
    fit = b0 + b1 * year
    
    m_y = np.average(ratio) 
    decadal_trend = 1000 * b1 / m_y
    
    ####### HYP TEST ########
    N = len(ratio) # Sample size
    rho1 = sm.tsa.acf(ratio)[1] # Lag-1 autocorrelation
    N_eff = N * ( (1 - rho1) / (1 + rho1) ) # Effective sample size
    
    # Standard errors
    x_err = ((year - year.mean()) ** 2).sum()
    y_err = ((ratio - fit)**2).sum()
    
    se = ((1 / (N_eff - 2)) * (y_err / x_err))**0.5
    
    # Hypothesis test
    conf = 0.95
    conf_adj = (1 + conf) / 2
    t_crit = stats.t.ppf(conf_adj, N_eff - 2) # Critical t-value
    t = b1 / se
    
    p = 2 * (1 - stats.t.cdf(np.abs(t), N_eff-2))
    
    ########################
    if decadal_trend > 0.05:
        trend_string = r'$+$%.1f percent per decade' % (decadal_trend)
        tc = 'red'
    elif decadal_trend < -0.05:
        trend_string = r'$-$%.1f percent per decade' % (-decadal_trend)
        tc = 'blue'
    else:
        trend_string = r'%.1f percent per decade' % (decadal_trend)
        tc = 'gray'


    # Plot timeseries
    plt.plot(year,ratio, 'k-', label='Precip fraction')
    plt.plot(year,stderr, 'k--', label='Std error')
    plt.plot(year,fit, '--', color=tc, ) #label='Unweighted fit'
    
    plt_title = "%s %s %s [%s–%s, %s–%s]" % (figlab[ibox],season[iseason], boxlabels[ibox],s0,s1,s2,s3)
    plt.title(plt_title, loc='left')
    
    
    if ibox==0 or ibox==3 or ibox==6:
        plt.ylabel("Fraction")
    if ibox>=6:
        plt.xlabel("Year")
    
    # Annotate fits
    plt.annotate("Mean: %.3f" % m_y, (1960,0.4*m_y), color=tc, fontsize=12)
    plt.annotate(trend_string, (1960,0.4*m_y-(0.3*0.4*m_y)), color=tc, fontsize=12)
    plt.annotate(rf'$p=${p:.3f}', (1960,0.4*m_y-(0.6*0.4*m_y)), color=tc, fontsize=12 )
    
    # Save trends
    timeseries.append((year, ratio, decadal_trend, fit, stderr, p, tc, plt_title, m_y, trend_string))

    if ibox == 0 : plt.legend()
    plt.grid()
    plt.ylim(bottom=0)
        
plt.subplots_adjust(hspace=0.25,wspace=0.2)
plt.tight_layout()

if export:
    plt.savefig(join(figure_dir, 'timeplots%s.pdf'%season[iseason]))
else:
    plt.show()
```
